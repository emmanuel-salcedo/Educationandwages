```python
import pandas as pd
import requests
from io import StringIO
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

Define the Function to Fetch Data from FRED API

```python
def fetch_fred_data(api_key, series_id, start_year, end_year):
    url = f'https://api.stlouisfed.org/fred/series/observations?series_id={series_id}&api_key={api_key}&file_type=json&observation_start={start_year}-01-01&observation_end={end_year}-12-31'
    for attempt in range(3):  # Retry up to 3 times
        try:
            response = requests.get(url)
            response.raise_for_status()  # Check for HTTP errors
            data = response.json()  # Parse the JSON response
            df = pd.DataFrame(data['observations'])  # Convert observations to DataFrame
            df['value'] = pd.to_numeric(df['value'], errors='coerce')  # Convert values to numeric
            return df
        except requests.exceptions.HTTPError as http_err:
            print(f"HTTP error occurred: {http_err}")
            if attempt < 2:
                print("Retrying...")
                time.sleep(5)  # Wait for 5 seconds before retrying
            else:
                print("Failed after 3 attempts.")
        except Exception as err:
            print(f"Other error occurred: {err}")
            break
```

### Normalization Process Explanation

Normalization is a technique used to scale data to a range, typically [0, 1], making it easier to compare datasets with different units or magnitudes. The process involves adjusting the values in the dataset so that they fit within a specified range, maintaining the relationships between the data points.

### Steps for Normalization

1. **Identify the Range of Values**:
   - For each column that needs to be normalized, determine the minimum and maximum values.
   
2. **Apply Min-Max Normalization**:
   - The min-max normalization formula is:
     \[
     \text{Normalized Value} = \frac{\text{Value} - \text{Min Value}}{\text{Max Value} - \text{Min Value}}
     \]
   - This formula transforms each value in the dataset to a corresponding value within the range [0, 1].

### Benefits of Normalization

- **Comparability**:
  - Normalization allows different datasets to be compared on the same scale, which is particularly useful when combining data with varying units or magnitudes.

- **Improved Performance**:
  - Many machine learning algorithms perform better when the input data is normalized, as it ensures that the model treats all features equally.


By normalizing the wages, inflation, and GDP data, we were able to bring all values to a common scale, enabling a more straightforward comparison and visualization. This process is essential for combining datasets and performing meaningful analyses across different types of data.

```python
# Normalize the wage columns
wage_columns = ['less_than_hs', 'high_school', 'some_college', 'bachelors_degree', 'advanced_degree']
wages_data[wage_columns] = (wages_data[wage_columns] - wages_data[wage_columns].min()) / (wages_data[wage_columns].max() - wages_data[wage_columns].min())

# Normalize the inflation data
inflation_data['value'] = (inflation_data['value'] - inflation_data['value'].min()) / (inflation_data['value'].max() - inflation_data['value'].min())

# Normalize the GDP data
gdp_data['value'] = (gdp_data['value'] - gdp_data['value'].min()) / (gdp_data['value'].max() - gdp_data['value'].min())

# Ensure the 'year' column is in datetime format
wages_data['year'] = pd.to_datetime(wages_data['year'], format='%Y')
inflation_data['date'] = pd.to_datetime(inflation_data['date'])
gdp_data['date'] = pd.to_datetime(gdp_data['date'])

# Rename 'date' columns to 'year' for consistency
inflation_data.rename(columns={'date': 'year'}, inplace=True)
gdp_data.rename(columns={'date': 'year'}, inplace=True)

# Merge the datasets on the 'year' column
merged_df = wages_data.merge(inflation_data[['year', 'value']], on='year', how='inner', suffixes=('', '_inflation'))
merged_df = merged_df.merge(gdp_data[['year', 'value']], on='year', how='inner', suffixes=('_inflation', '_gdp'))

# Plot the normalized data
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
sns.lineplot(data=merged_df, x='year', y='less_than_hs', label='Less than High School')
sns.lineplot(data=merged_df, x='year', y='high_school', label='High School')
sns.lineplot(data=merged_df, x='year', y='some_college', label='Some College')
sns.lineplot(data=merged_df, x='year', y='bachelors_degree', label="Bachelor's Degree")
sns.lineplot(data=merged_df, x='year', y='advanced_degree', label='Advanced Degree')
sns.lineplot(data=merged_df, x='year', y='value_inflation', label='Inflation', color='red')
sns.lineplot(data=merged_df, x='year', y='value_gdp', label='GDP', color='green')
plt.title('Normalized Wages by Education Level, Inflation, and GDP')
plt.xlabel('Year')
plt.ylabel('Normalized Values')
plt.legend()
plt.show()
```
