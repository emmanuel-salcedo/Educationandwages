```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
```

```python
# MySQL connection details
mysql_user = 'root'
mysql_password = 'phones123'
mysql_host = 'localhost'  # or your database IP
mysql_port = '3306'  # default MySQL port
mysql_db = 'EducationWage'
```

```python
# SQLAlchemy engine for MySQL connection
engine = create_engine(f'mysql+mysqlconnector://{mysql_user}:{mysql_password}@{mysql_host}:{mysql_port}/{mysql_db}', echo=False)
mysql_table = 'educationandwages'

# Path to your CSV file
wages_data = 'wages_by_education.csv'

# Read CSV into DataFrame
df = pd.read_csv(wages_data)
df.head()
```

```python
# Write the DataFrame to the specified table in MySQL
df.to_sql(name=mysql_table, con=engine, if_exists='replace', index=False)

print(f"Data from {wages_data} has been successfully loaded into the {mysql_table} table in the {mysql_db} database.")
```

```python
# Select only the relevant columns
education_levels = ['less_than_hs', 'high_school', 'some_college', 'bachelors_degree', 'advanced_degree']
years = df['year']

# Plot the trends
plt.figure(figsize=(12, 6))
for level in education_levels:
    plt.plot(years, df[level], label=level.replace('_', ' ').title())

plt.xlabel('Year')
plt.ylabel('Median Wage ($)')
plt.title('Wage Trends Across Education Levels Over Time')
plt.legend(title='Education Level')
plt.grid(True)
plt.show()
```

```python
# Plot 2: Gender wage gap across educational levels
men_data = df[['year', 'men_less_than_hs', 'men_high_school', 'men_some_college', 'men_bachelors_degree', 'men_advanced_degree']]
women_data = df[['year', 'women_less_than_hs', 'women_high_school', 'women_some_college', 'women_bachelors_degree', 'women_advanced_degree']]

men_data.columns = ['year', 'less_than_hs', 'high_school', 'some_college', 'bachelors_degree', 'advanced_degree']
women_data.columns = ['year', 'less_than_hs', 'high_school', 'some_college', 'bachelors_degree', 'advanced_degree']

fig, ax = plt.subplots(2, 1, figsize=(12, 10), sharex=True)

for level in education_levels:
    ax[0].plot(years, men_data[level], label=level.replace('_', ' ').title())
    ax[1].plot(years, women_data[level], label=level.replace('_', ' ').title())

ax[0].set_title('Men Wages Across Education Levels')
ax[1].set_title('Women Wages Across Education Levels')

for axis in ax:
    axis.set_ylabel('Median Wage ($)')
    axis.legend(title='Education Level')
    axis.grid(True)

ax[1].set_xlabel('Year')
plt.tight_layout()
plt.show()
```

```python
# Overlay men vs women wage gap across education levels

# Calculate wage gap
gender_gap = men_data.set_index('year') - women_data.set_index('year')
gender_gap = gender_gap.reset_index()

# Plot the overlay
plt.figure(figsize=(12, 6))
for level in ['less_than_hs', 'high_school', 'some_college', 'bachelors_degree', 'advanced_degree']:
    plt.plot(gender_gap['year'], gender_gap[level], label=level.replace('_', ' ').title())

plt.xlabel('Year')
plt.ylabel('Wage Gap ($)')
plt.title('Wage Gap Between Men and Women Across Education Levels (Overlay)')
plt.legend(title='Education Level')
plt.grid(True)
plt.show()
```
