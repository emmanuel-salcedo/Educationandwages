```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sqlalchemy import create_engine
```

```python
# MySQL connection details
mysql_user = 'root'
mysql_password = 'phones123'
mysql_host = 'localhost'  # or your database IP
mysql_port = '3306'  # default MySQL port
mysql_db = 'EducationWage'
```

```python
# SQLAlchemy engine for MySQL connection
engine = create_engine(f'mysql+mysqlconnector://{mysql_user}:{mysql_password}@{mysql_host}:{mysql_port}/{mysql_db}', echo=False)
mysql_table = 'educationandwages'

# Path to your CSV file
csv = 'wages_by_education.csv'

# Read CSV into DataFrame
df = pd.read_csv(csv)
df.head()
```

```python
# Write the DataFrame to the specified table in MySQL
df.to_sql(name=mysql_table, con=engine, if_exists='replace', index=False)

print(f"Data from {csv} has been successfully loaded into the {mysql_table} table in the {mysql_db} database.")
```
